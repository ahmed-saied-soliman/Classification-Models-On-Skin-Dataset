```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
```

SVM Implmentation


```python
data=pd.read_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/train_all.csv")
data_test=pd.read_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/test_all.csv")
x_train=data.iloc[:,1:4]
y_train=data.iloc[:,4:]
training = svm.SVC()
training.fit(x_train,y_train.values.ravel())
x_test=data_test.iloc[:,1:]
y_test=training.predict(x_test)
print(pd.DataFrame(y_test))
result=pd.DataFrame(y_test)
result.to_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/resultsvm.csv")
```
