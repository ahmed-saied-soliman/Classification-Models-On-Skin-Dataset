```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

# Plotting Features Of Distribution

```python
train_data=pd.read_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/train_all.csv")
test_data=pd.read_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/test_all.csv")
y_test_DF=pd.read_csv("C:/Users/Ahmed Saied/Downloads/Compressed/knn/C_result_svm.csv")
for i in range(1,len(train_data.columns)-1):
    print("Scatter Plot Between"+" "+train_data.columns[i]+" & Class")
    plt.scatter(train_data.iloc[:,i],train_data.iloc[:,-1])
    plt.show()
for i in range(1,len(test_data.columns)):
    print("Scatter Plot Between"+" "+test_data.columns[i]+" & Class")
    plt.scatter(test_data.iloc[:,i],y_test_DF.iloc[:,-1])
    plt.show()
```

# Correlation Between Features

```python
for i in range(1,len(train_data.columns)-1):
    for j in range (1,len(train_data.columns)-1):
        print("Scatter Plot Between"+" "+train_data.columns[i]+" & "+train_data.columns[j])
        plt.scatter(train_data.iloc[:,i],train_data.iloc[:,j])
        plt.show()
for i in range(1,len(test_data.columns)):
    for j in range (1,len(test_data.columns)):
        print("Scatter Plot Between"+" "+test_data.columns[i]+" & "+test_data.columns[j])
        plt.scatter(test_data.iloc[:,i],test_data.iloc[:,j])
        plt.show()
corr=train_data.iloc[:,1:4].corr()
sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns)
corr_test=test_data.iloc[:,1:].corr()
sns.heatmap(corr_test,xticklabels=corr.columns,yticklabels=corr.columns)
```
